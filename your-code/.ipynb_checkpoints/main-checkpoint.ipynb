#### Display the trending Python repositories in GitHub

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
#your code

html=requests.get(url)

print(html)

soup=BeautifulSoup(html.text,"lxml")

print(soup)
````

````python
#--Make a list all the repos of the page

repos=soup.select('h1[class="h3 lh-condensed"] >a') #selects al the tags with given characteristics


repos_clean=[element.text.strip() for element in repos] #extract text and strips the text from the tags


print(repos_clean)




````

#### Display all the image links from Walt Disney wikipedia page

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/Walt_Disney'
````

````python
#your code
html=requests.get(url)

print(html)

soup=BeautifulSoup(html.text,'lxml')

#soup=soup.prettify()

print(soup)
````

#### Number of Titles that have changed in the United States Code since its last release point 

````python
# This is the url you will scrape in this exercise
url = 'http://uscode.house.gov/download/download.shtml'

````

````python
#your code

html=requests.get(url)

print(html)

souptitles=BeautifulSoup(html.text,'lxml')

print(souptitles)
````

````python


counting=souptitles.select('div.usctitle')  #[class="usctitle"]

listing=[element.text.strip() for element in counting]


del listing[0:2]
print("there has been ",len(listing)," changes.")


````
